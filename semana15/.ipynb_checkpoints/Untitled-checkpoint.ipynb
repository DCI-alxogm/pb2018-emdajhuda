```python
#aqui importe las librerias
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as Sc

%matplotlib inline
```

```python
#con esta funcion se lee la matriz del archivo de texto
inputs=np.loadtxt('IRCF.txt') 
```

```python
#Selecciono las columnas 1,2 y 4 y las guardo en diferentes variables
f=inputs[:,0:1]
e=inputs[:,1:2]
I=inputs[:,3:4]
#Hago un arreglo de la temperatura para que lo pueda evaluar de 1 a 5 kelvin
t=np.linspace(1,5,100000)
#las variables x son para el error
x=0
xg=4000
#inicializo la temperatura del universo a 0
tu=0
xm=np.linspace(0,0,len(I))
#g lo utilizo para graficar los errores
g=np.linspace(0,0,len(I))
```

```python
#Asigo los valores de las constantes
h=Sc.Planck
k=Sc.Boltzmann
c=Sc.c
```

```python
#con subplots indico que voy a juntar dos graficas, la de los datos del experimento con sus errores
fig, axs = plt.subplots(2, 1, sharex=True)
fig.subplots_adjust(hspace=0)
axs[0].plot(f, e)
axs[1].errorbar(f,g,I)
```

```python
#Se define la funcion dada para calcular el espectro de la radiacion de un cuerpo negro y al final le agrego una multiplicacion para cambiar a Jy
def i(f,t):
    return ((h*(pow(f,3)))/(pow(c,2)))*(1/((np.exp((h*f)/(k*t)))-1))*(10**20)
```

```python
#evaluo la funcion con muchas temperaturas
for w in range(len(t)):
    x=0
    J=i(f*c*10**2,t[w])
#Encuentro el error
    for z in range(len(I)):
        xm[z]=((e[z]-J[z])/I[z])**2
        x=x+xm[z]
#Si el error es mas pequeo se guarda ese error para la proxima interacion y guarda esa temperatura como la temperatura del universo
    if (x<xg):
        xg=x
        tu=t[w]
#Le digo que grafique cada 10,000 temperaturas 
    if (w%10000==0):
        plt.plot(f,J,color='y')

#Aqu grafico con la temperatura del universo y la comparo con la de los datos del experimento
print(tu)
J=i(f*c*10**2,tu)
plt.scatter(f,e)
plt.plot(f,J)
plt.xlabel('Frecuencia $cm^{-1}$')
plt.ylabel('Espectro MJy/sr')
plt.legend()
```

```python
#En esta parte compare los datos del experimento, la temperatura que obtuve y la temperatura del universo que encontre en internet (2.792 K)
J=i(f*c*10**2,tu)
plt.plot(f,J)
J=i(f*c*10**2,2.792)
plt.plot(f,J)
plt.scatter(f,e)
plt.xlabel('Frecuencia $cm^{-1}$')
plt.ylabel('Espectro MJy/sr')
plt.legend()
```
