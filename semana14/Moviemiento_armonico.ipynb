# Movimiento armonico
### Declaración de librerias

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Definición de funciones

```python
def velocidad(v,h,T,O):
    return v+(-h*((2*np.pi/T)**2)*O)
def angulo(O,h,v):
    return O+h*v
```

### Inicialización de variables y creación de arrays

```python
i=1000000
d=20
T=4
O=np.linspace(180,180,i)
h=d/i
t=np.linspace(0,d,i)
v=np.linspace(0,0,i)
```

### Aquí se andan llamar las funciones y se calculan la velocidad y angulo con respecto al tiempo

```python
for j in range(i-1):
    v[j+1]=velocidad(v[j],h,T,O[j])
    O[j+1]=angulo(O[j],h,v[j])
```

### Gráfica de la velocidad con respecto al tiempo

```python
plt.plot(v,O, color='r', label='Velocidad con respecto al angulo')
plt.xlabel('v')
plt.ylabel('O')
plt.legend()
```

### Gráfica del tiempo con respecto a la velocidad

```python
plt.plot(t,v, color='r', label='Tiempo con respecto a velocidad')
plt.xlabel('t')
plt.ylabel('v')
plt.legend()
```

### Gráfica del tiempo con respecto al angulo

```python
plt.plot(t,O, color='r', label='Tiempo con respecto al angulo')
plt.xlabel('t')
plt.ylabel('O')
plt.legend()
```
