```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Opción 1 para definir funciones
fx=lambda x: (x**2)*np.cos(x)-3*x**3
```

```python
x=np.linspace(-10,10,100)
x2=np.linspace(0,5,50)
y=(x**2)*np.cos(x)-3*x**3
y2=(x2**2)*np.cos(x2)-3*x2**3
```

```python
plt.plot(x,y)
plt.plot(x2,y2)
```

```python
plt.plot(x2,y2)
```

```python
fx=lambda x,y: x*y*np.cos(x/y)-3*x**3
```

```python
x=np.linspace(-10,10,100)
y=np.linspace(0,5,100)
```

```python
plt.plot(x,fx(x,y))
```

```python
#Opción 2: para definir funciones.
def fx2(x,y):
    z=x*y*np.cos(x/y)-3*x**3
    return z
```

```python
x=np.linspace(-10,10,100)
y=np.linspace(0,5,100)
z=fx2(x,y)
plt.plot(x,z)
```

```python
def recta(a,m,b=0):
    return m*a+b
```

```python
x=np.linspace(-5,5,20)
plt.plot(x,recta(x,3), color='y')
plt.plot(x,recta(x,-3), color='y')
plt.plot(x,recta(x,3,-2), color='r')
plt.plot(x,recta(x,-3,-2), color='r')
plt.plot(x,recta(x,3,2), color='g')
plt.plot(x,recta(x,-3,2), color='g')
plt.grid()
```

```python
def recta(a,m=1,b=0,label='x'):
    z=m*a+b
    plt.plot(a,z,label=label)
    return
```

```python
recta(x)
recta(x,-2,1,'-x+1')
plt.legend()
plt.grid()
```
